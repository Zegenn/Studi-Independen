## Intro to AI-1

1. Apa itu AI?
   Jawab:
   Kecerdasan yang ditambahkan kepada suatu sistem yang bisa diatur dalam konteks ilmiah atau bisa disebut juga inteligensi artifisial atau hanya disingkat AI.

2. Apa bedanya AI, ML dan DL?
   Jawab:
   - Artificial Intellegence(AI) : Teknik apa pun yang memungkinkan komputer meniru kecerdasan manusia.
   - Machine Learning(ML) : Bagian dari AI yang menggunakan data dan pelatihan untuk menemukan pola/model yang membantu mesin untuk mengerjakan tugasnya lebih baik.
   - Deep Learning(DL) : Bagian dari ML yang menggunakan jaringan saraf tiruan untuk membuat mesin belajar sendiri dalam memilih fitur untuk meningkatkan performanya.

3. Apa 3 level di AI?
   Jawab:
   - Artificial Narrow Intelligence(ANI) : AI yang melakukan 1 atau sedikit kemampuan tertentu. Misalnya menjawab pertanyaan (AI chatbot), face recognition, machine translation, delivery robot, dll.
   - Artificial General Intelligence(AGI) : AI yang dapat melakukan banyak hal layaknya manusia.
   - Artificial Super Intelligence(ASI) : Tahap dimana AI sudah self-conscious, memiliki kesadaran sendiri.


## Intro to AI-2

1. Sebutkan urutan piramida DIKW dan penjelasannya!
   Jawab:
   - Wisdom
     Wisdom sebagai pemahaman yang dievaluasi dimana pemahaman mengemukakan diskrit Knowledge dan Wisdom.
   - Knowledge
     Knowledge merujuk pada informasi yang telah di proses terorganisir atau terstruktur dalam beberapa cara apabila diterapkan.
   - Information
     Information didefinisikan sebagai data yang mengandung makna dan tujuan setelah diolah.
   - Data
     Fakta dan keterangan yang mendeskripsikan objek.

2. Apa bedanya structured data dan unstructured data?
   Jawab:
   - Structured  data : Data dengan skema yang jelas sehingga mudah untuk dianalisa maupun diintegrasikan dengan data terstruktur lainnya.
   - Unstructured data : Data dengan berbagai bentuk sehingga sangat sulit untuk dianalisis maupun diintegrasikan dengan sumber data lain.

3. Jelaskan 3 domain di AI!
   Jawab:
   - Data Science = Ilmu interdisipliner yang menggabungkan teori di ilmu komputer, matematika, serta bidang ilmu tertentu untuk mengolah data menjadi sebuah informasi dan wawasan.
   - Natural Language Processing = Domain di AI dimana datanya berupa teks dan suara.
   - Computer Vision = Domain AI dimana datanya berupa gambar dan vidio.

## ML-4

1. Apa yang dimaksud Logistic Regression?
   Jawab:
   Logistic Regression adalah salah satu metode dalam statistika yang digunakan untuk melakukan analisis regresi pada data dengan variabel target yang bersifat biner, yaitu memiliki dua kelas atau nilai saja.
2. Bagaimana cara kerja Logistic Regression?
   Jawab:
   - Logistic regression mengukur hubungan antara variabel target (yang ingin diprediksi) dan variabel input (fitur yang digunakan) menggunakan fungsi logistik.
   - Probabilitas dihitung menggunakan fungsi sigmoid, sehingga dapat mengubah nilai-nilai yang diperoleh  menjadi 0 atau 1.
3. Apa yang dimaksud Support Vector Machine(SVM)?
   Jawab:
   Metode dalam Supervised learning yang biasanya digunkaan untuk klasifikasi (seperti Support Vector Classification) dan regresi (Support Vector Regression).
4. Sebutkan kelebihan dan kekurangan dari SVM?
   Jawab:
   - Kelebihan:
     a. Klasifikasi menggunakan model SVM memiliki akurasi yang baik pada data berdimensi tinggi.
     b. Membutuhkan memori yang sedikit.
   - Kekurangan:
     a. Walaupun memperbolehkan beberapa observasi boleh tidak berada pada ruang yang tepat (berdasarkan margin), support vector classifier masih memiliki asumsi bahwa decision boundary berbentuk suatu fungsi linear. Ini adalah batasan utama model ini.
     b. Klasifikasi menggunakan SVM tidak berfungsi dengan baik untuk kelas yang saling tumpang tindih.
5. Kernel apa saja yang mampu meningkatkan performa SVM?
   Jawab:
   - Linear Kernel
   - Polynomial Kernel
   - Radial Basis Function Kernel (RBF)/ Gaussian Kernel
6. Apa saja Evaluasi Metrices yang bisa digunakan untuk kasus klasifikasi?
   Jawab:
   - Akurasi (Accuracy)
   - Presisi (Precision)
   - Recall (Sensitivitas atau True Positive Rate)
