2. Apa itu Hierarchial Clustering?

   Jawab:
   
   Data dikelompokan melalui suatu bagan yang berupa hirarki, dimana terdapat penggabungan dua grup yang terdekat di setiap iterasinya ataupun pembagian dari seluruh set data ke dalam cluster.

   Jawab:

   Representasi grafis dari hierarki atau struktur berhirarki dari sebuah set data.

1. Jelaskan mengenai Python Web Framework (jenis dan kelebihannya)?

   Jawab:
   
   Jenis:

   a. Django = salah satu framework web Python yang paling populer dan kuat. Ia menyediakan berbagai fitur bawaan termasuk ORM (Object-Relational Mapping), sistem admin otomatis, routing URL, manajemen sesi, dan keamanan.

   b. Flask = framework web Python yang lebih ringan dan fleksibel dibandingkan dengan Django. Ia menyediakan dasar-dasar yang diperlukan untuk membangun aplikasi web, tetapi memungkinkan pengembang untuk memilih sendiri alat dan pustaka yang ingin digunakan.

   c. FastAPI = framework web Python yang cepat dan modern untuk membuat API dengan mudah. Ia menggunakan pendekatan berbasis deklaratif dan menawarkan kecepatan yang sangat tinggi berkat integrasinya dengan Pydantic untuk validasi data dan Starlette untuk manajemen permintaan HTTP.

   Kelebihan:

   a. Framework-web Python menyediakan banyak alat bawaan dan konvensi yang mempercepat pengembangan aplikasi web.

   b. Python Web Framework seperti Django dan Flask dirancang untuk mendukung pembangunan aplikasi web yang mudah diperluas dan diskalakan.

   c. Framework-web Python dirancang untuk mengikuti prinsip-prinsip Pythonic, yang berarti mereka memiliki sintaks dan pola desain yang mudah dipahami dan dipelajari oleh pengembang Python.

2. Jelaskan mengenai Flask (Fitur, Kelebihan dan kekurangan)!

   Jawab:

   Fitur:

   a. Routing = Routing adalah URL yang digunakan untuk memandu pengunjung ke berbagai bagian web app.

   b. Template Rendering

   c. Extensions: Flask memiliki ekosistem yang luas dari ekstensi (extensions) yang memungkinkan pengembang untuk menambahkan fungsionalitas tambahan ke aplikasi mereka.

   Kelebihan:

   a. Flask adalah salah satu framework web Python yang paling ringan dan fleksibel, yang memungkinkan pengembang untuk membangun aplikasi sesuai dengan kebutuhan mereka tanpa harus memuatkan fitur-fitur yang tidak diperlukan.

   b. Flask memiliki kurva belajar yang sangat landai, membuatnya sangat mudah dipelajari oleh pengembang pemula atau yang memiliki pengalaman terbatas dalam pengembangan web.

   Kekurangan:

   a. Salah satu kekurangan Flask adalah kurangnya fitur bawaan jika dibandingkan dengan framework web seperti Django.

   b. Karena sifatnya yang minimalis, Flask seringkali bergantung pada ekstensi untuk menambahkan fungsionalitas tambahan.

3. Bagaimana cara Flask bekerja?

   Jawab:

   a. Pertama-tama, harus membuat objek Flask sebagai aplikasi web Anda. Ini dilakukan dengan membuat sebuah instance dari kelas Flask.

   b. Setelah aplikasi dibuat, dapat menentukan routing URL dengan menggunakan dekorator. Routing URL menentukan bagaimana aplikasi menanggapi permintaan yang diterima.

   c. Setiap routing URL dihubungkan dengan sebuah fungsi, yang biasa disebut dengan View Function. Fungsi ini akan dijalankan ketika URL yang sesuai diakses oleh pengguna.

   d. Setelah aplikasi dikonfigurasi, kita perlu menjalankannya. Kita dapat menggunakan metode run() yang disediakan oleh objek aplikasi.

   e. Ketika aplikasi dijalankan, ia akan mulai mendengarkan permintaan HTTP dari klien (browser atau aplikasi lain). Ketika permintaan diterima, Flask akan memeriksa routing URL untuk menemukan fungsi yang sesuai.
